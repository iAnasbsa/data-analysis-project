## Investigation Overview


Bay Wheels (previously known as Ford GoBike) is a nearby public motorcycle sharing gadget within the San Francisco Bay vicinity, California. Bay Wheels is the primary nearby and big-scale bicycle sharing system deployed in California and at the West Coast of the united states with nearly 500,000 rides for the reason that release in 2017 and had about 10,000 annual subscribers as of January 2018. This exploration and visualization examine basically focuses on locating the bike usage sample and consumer dependancy traits from the 12-month ride journeys information in 2018.


## Dataset Overview

The dataset used for this exploratory evaluation includes monthly character ride data from January 2018 to December 2018 in CSV layout masking the greater San Francisco Bay region, raw data is available here. Visualizations below are comprised of wrangled and cleaned information to facilitate exploration evaluation and assist find out utilization sample and rider traits.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
all_df = pd.read_csv(r'D:\Downloads\Last project\Project Template\alloutput.csv')
```

###  Trend over the Year for the trips
##### The carrier is really affected by the seasons, so the usage of the service decreases in iciness and will increase in other seasons..

```python
all_df.groupby('starttime').agg({'bikeid':'count'}).plot( legend=False, figsize=(21,11))
plt.title('Trips trend over the year', fontsize=28)
plt.xlabel('date', labelpad=28)
plt.ylabel('Count', labelpad=28)
```

###  Customer trip duration (min)
##### most of tripst under 1 hour and usually it's around 5 to 20 min ..

```python
plt.hist(all_df['tripduration'],bins=range(0,90))
plt.xlabel('min');
plt.title('Customer Trip Duration in min', fontsize=19)
plt.ylabel('Count (M)');
```

###  Age  Distribution
##### for age, we can see that the distribution is between 30 to 45 years old.


```python
all_df_age = all_df.loc[all_df['age']<=90]
```

```python
plt.figure(figsize=(18, 12))
bin_edges = np.arange(0, all_df_age.age.max()+2.1, 2.1)
plt.hist(data=all_df_age, x='age', bins=bin_edges);
plt.xlim(13, 90);
plt.xlabel('age');
plt.title('Customer age distribution', fontsize=19)
plt.ylabel('Count (M)');
```
